```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
%matplotlib inline

from sklearn.feature_extraction.text import CountVectorizer
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.max_colwidth', -1)

import re
import string
from textblob import TextBlob
import seaborn as sns

plt.style.use('ggplot')
font = {'weight' : 'bold',
        'size'   : 14}
plt.rc('font', **font)
```

```python
df = pd.read_pickle('/Users/ellykyles/capstone2/data/train.pkl')
```

```python
pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity

df['polarity'] = df['clean_text'].apply(pol)
df['subjectivity'] = df['clean_text'].apply(sub)
```

```python
plt.subplots(figsize=(7,7), dpi=80)
sns.distplot(df.loc[df.Score==1, "polarity"] , color="dodgerblue", label="1")
sns.distplot(df.loc[df.Score==2, "polarity"] , color="yellow", label="2")
sns.distplot(df.loc[df.Score==3, "polarity"] , color="orange", label="3")
sns.distplot(df.loc[df.Score==4, "polarity"] , color="palevioletred", label="4")
sns.distplot(df.loc[df.Score==5, "polarity"] , color="deeppink", label="5")

plt.title('Polartiy_Distribution')

plt.legend();
plt.savefig('Polarity_distribution')
```

```python
plt.subplots(figsize=(7,7), dpi=80)
sns.distplot(df.loc[df.Score==1, "words_count"] , color="dodgerblue", label="1")
sns.distplot(df.loc[df.Score==2, "words_count"] , color="yellow", label="2")
sns.distplot(df.loc[df.Score==3, "words_count"] , color="orange", label="3")
sns.distplot(df.loc[df.Score==4, "words_count"] , color="palevioletred", label="4")
sns.distplot(df.loc[df.Score==5, "words_count"] , color="orchid", label="5")

plt.title('Word Count Distribution')
plt.xlim((0,500))
plt.legend();
plt.savefig('word_counts_distribution')
```

```python
plt.subplots(figsize=(7,7), dpi=80)
sns.distplot(df.loc[df.Score==1, "subjectivity"] , color="dodgerblue", label="1")
sns.distplot(df.loc[df.Score==2, "subjectivity"] , color="yellow", label="2")
sns.distplot(df.loc[df.Score==3, "subjectivity"] , color="orange", label="3")
sns.distplot(df.loc[df.Score==4, "subjectivity"] , color="palevioletred", label="4")
sns.distplot(df.loc[df.Score==5, "subjectivity"] , color="deeppink", label="5")

plt.title("Subjectivity_Distribution")

plt.legend();
plt.savefig('Subjectivity_Distribution')
```
